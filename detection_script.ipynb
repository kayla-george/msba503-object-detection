```python
from pathlib import Path 
import os 

#1. Tell Python where images live
IMAGE_DIR = 'Photos'

#2. Turn that into a Path object
image_dir = Path(IMAGE_DIR)

#3. Safety check: does the folder exist? 
if not image_dir.exists():
    raise FileNotFoundError(f"Folder '{IMAGE_DIR}' not found. Check your path/name.")

#4. Collect all image files with common extensions
valid_exts = {".jpg", ".jpeg", ".png", ".bmp"}
image_paths = [p for p in image_dir.iterdir() if p.suffix.lower() in valid_exts]

#5. Print a quick summary
print(f"Found {len(image_paths)} images in '{IMAGE_DIR}':")
for p in image_paths:
    print(" -", p.name)
```
